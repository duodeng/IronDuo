```python
# Cleaning column gender using map function and another user defined function

lab_Customer_df['gender'].unique()
```

```python
# clean the gender
def clean_gender(x):
    if x in ['M', 'Male']:
        return 'Male'
    elif x in ['F', 'female','Female']:
        return 'Female'
    elif np.nan:  pass
    else:
        return 'U'
lab_Customer_df['gender'] = list(map(clean_gender, lab_Customer_df['gender'])) 
lab_Customer_df
```
