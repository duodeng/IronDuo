### 3. Removing outliers using 1.5*IQR technique for all numerical columns.

```python
for col in lab_Customer_df.columns:
    print(f"{col} : {lab_Customer_df[col].dtype}")     
```

```python
# The IQR is the difference between Q3 and Q1(Q3-Q1)
# calculate the Q1 and Q3
for column in lab_Customer_df.select_dtypes(include=['float64']).columns:
    Q3 = lab_Customer_df[column].quantile(0.75)
    Q1 = lab_Customer_df[column].quantile(0.25)
    IQR = Q3 - Q1   
# determine the lower bound and upper bound
    low_bound = Q1 - (1.5*IQR)
    upper_bound = Q3 + (1.5*IQR)
    
    outliers= lab_Customer_df[(lab_Customer_df[column] < low_bound) | (lab_Customer_df[column] > upper_bound)]
    print(f'From { low_bound } to { upper_bound} for {column}')
# delete the outliers
lab_Customer_df = lab_Customer_df.drop(outliers.index)

# he dataset now contains only rows without outliers
lab_Customer_df
```
