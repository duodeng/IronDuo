## 1. getting the numeric data into dataframe and categorical columns

```python
# getting numrical columns and categorical columns
numericals = data.select_dtypes(include=[np.number])
categoricals = data.select_dtypes(include=[np.object])
```

## 2. checking the normality of the numerical variables visually

```python
all_col_num = data.select_dtypes(include=['int64']).columns.tolist()
all_col_num 
```

```python
# 1. Use Matplotlib to construct histograms.
fig, axs = plt.subplots(2,4,figsize = (16,10))
for i,var in enumerate(all_col_num):
    ax = axs.ravel()[i] # ravel convert 2d to 1d array
    
    ax.hist(numericals[var], bins=15)
    ax.set_xlabel(var) #
    ax.set_ylabel('Frequency')
    ax.set_title(f' {var}')

# add a title for the whole figure
fig.suptitle('Data Histograms')

# display the plot
plt.show()
```

```python
# 2. Use seaborn library to construct distribution plots for the numerical variables
fig, axes = plt.subplots(2, 4, figsize=(15, 10)) # creates a 2x4 grid of subplots
axes = axes.ravel()  # flatten the multi-dimensional array into a one-dimensional array,then 0,1,2...7

for i, var in enumerate(all_col_num):
    ax = axes[i]
    sns.distplot(numericals[var], ax=ax) # ax=ax specifies the axis on which the plot should be drawn
    ax.set_title(f'{var}') # sets the title for each subplot

plt.show() # plt.tight_layout() 
```

## 5.2 The relationship between categorical columns and total claim amount

```python
categorical_col = data.select_dtypes(include = object).columns # select columns with data type= object
len(categorical_col) 

fig, axs = plt.subplots(5,3, figsize=(20,45) )# creat canves
axs = axs.ravel() # flatten a multi-dimensional array of "axs" into a 1-dimensional array. 
for i, var in enumerate(categorical_col):
    sum_total_claim_amount = data.groupby(var)['total_claim_amount'].sum()
    sns.barplot(data = data, x = sum_total_claim_amount.index, y = sum_total_claim_amount.values, palette = 'Blues', ax = axs[i])
    axs[i].set_xticklabels(axs[i].get_xticklabels(), rotation=30)
    axs[i].set_title(f'Total Claim Amount by {var}')
enumerate(categorical_col)    
```
