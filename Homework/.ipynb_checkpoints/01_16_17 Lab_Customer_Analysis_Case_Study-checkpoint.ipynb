## 8.  correcting typos and filtering data (state+gender)

```python
# 8.1 finding the unique values in the column 'st'.
df['st'].unique()
```

```python
# 8.2 correcting state typos
def clean_state(x):
    if x in ['AZ', 'Arizona']:
        return 'Arizona'
    elif x in ['WA', 'Washington']:
        return 'Washington'
    elif x in ['Cali','California']:
        return 'California'
    else:
        return x
df['st'] = list(map(clean_state, df['st'])) 
```

```python
# 8.3 finding the unique values in the column 'gender'.
df['gender'].unique()
```

```python
# 8.4 correcting gender typos
def clean_gender(x):
    if x in ['M', 'Male']:
        return 'Male'
    elif x in ['F', 'female','Female']:
        return 'Female' 
    else:
        return x
    
df['gender'] = list(map(clean_gender, df['gender'])) 
```

```python
# 8.5 standardize the texts in 'st', 'gender'columns
df['st'] = df['st'].str.lower()
df['gender'] = df['gender'].str.lower()
df
```

```python
# replace column "State" to different zones
def replace_state_with_zone():
    global df
    df['Zone'] = df['st']
    state_to_zone = {
        'california': 'West Region',
        'oregon': 'North West',
        'washington': 'East',
        'arizona': 'Central',
        'nevada': 'Central'
    }
    df['Zone'] = df['st'].map(state_to_zone)
# Drop the "st" column
    df = df.drop(columns=['st'])
    return df
replace_state_with_zone() 
```

### 1. nerging the two categories Luxury SUV and Luxury Car into one category

```python
def merge_luxury_categories():
# Replace all instances of "Luxury SUV" with "Luxury Vehicle"
    df['vehicle class'] = df['vehicle class'].replace("Luxury SUV", "Luxury Vehicle")
# Replace all instances of "Luxury Car" with "Luxury Vehicle"
    df['vehicle class'] = df['vehicle class'].replace("Luxury Car", "Luxury Vehicle")
    return df
merge_luxury_categories()
```

### 2. Removing outliers using 1.5*IQR technique for all numerical columns.

```python
# 1. check the information before we removing outliers.
df.describe().T
```

```python
# 2. calculate the outliers
for column in df.select_dtypes(include=['float64']).columns:
    Q3 = df[column].quantile(0.75)
    Q1 = df[column].quantile(0.25)
    IQR = Q3 - Q1   
# determine the lower bound and upper bound
    low_bound = Q1 - (1.5*IQR)
    upper_bound = Q3 + (1.5*IQR)
    
    outliers= df[(df[column] < low_bound) | (df[column] > upper_bound)]
    print(f'From { low_bound } to { upper_bound} for {column}')
    
# delete the outliers
df = df.drop(outliers.index)

# he dataset now contains only rows without outliers
df
```
