### Gathering insights:

- How many car makers are there? How many models? Which car maker has the most cars in the dataset?

- When were these cars made? How big is the engine of these cars?

- What's the frequency of different transmissions, drivetrains and fuel types?

- What's the car that consumes the least/most fuel?

```python
# 1.1 How many car makers are there? How many models? Which car maker has the most cars in the dataset?
num_makers= len(df_cars["Make"].unique())
print(f'There are {num_makers} makers.')

# 1.2 How many models?
num_models = len(df_cars["Model"].unique())
print(f'There are {num_makers} models.')

# 1.3 Which car maker has the most cars in the dataset?
most_car = df_cars["Make"].value_counts()
most_car_maker = most_car.idxmax() # idxmax get the name(index) of maximum value
print(f'{most_car_maker} has the most cars.')
```

```python
# 2.1 When were these cars made? 

most_car_filter = df_cars["Make"].str.startswith("Chevrolet") #str.startswith() check if a string starts with a specified substring.
df_cars.loc[most_car_filter, ["Make",'Model', 'Year']].sort_values(by="Year") # sort_values(by="") function can sort the column by name

# 2.2 How big is the engine of these cars?
most_car_filter = df_cars["Make"].str.startswith("Chevrolet")
df_cars.loc[most_car_filter, ["Make",'Model', 'Engine Displacement']].sort_values(by="Engine Displacement")
```

```python
# 3.What's the frequency of different transmissions, drivetrains and fuel types?
fre_trans = df_cars["Transmission"].value_counts()
```

```python
fre_dri = df_cars["Drivetrain"].value_counts()
```

```python
fre_types =  df_cars["Fuel Type"].value_counts()
```

```python
# 4.What's the car that consumes the least/most fuel?
rank_all = df_cars.sort_values(by="Fuel Barrels/Year",ascending= False)# ascending= False (high-low)
max_cost = rank_all["Fuel Barrels/Year"].max()
max_cost
min_cost = rank_all["Fuel Barrels/Year"].min()
min_cost

# consumes the most 
car_consume_most =rank_all.loc[ rank_all["Fuel Barrels/Year"] == max_cost]
car_consume_most
car_consume_most_name = car_consume_most["Make"].unique()[0] # [0]前面是一个array
car_consume_most_name
print(f'{car_consume_most_name} consumes the most fuel.')
```

```python
# consumes the least 
car_consume_least =rank_all.loc[ rank_all["Fuel Barrels/Year"] == min_cost]
car_consume_least
car_consume_least_name = car_consume_least["Make"].unique()[0] # [0]前面是一个array
car_consume_least_name
print(f'{car_consume_least_name} consumes the least fuel.')
```
